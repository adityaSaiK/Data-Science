The dataset is of MNIST images which are flattened. So the 28*28 image is flattened to form a 784 feature vector.

So, let's now load the data and separate the independent and dependent variables i.e., the features and labels.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.linalg import eigh
import seaborn as sns
```

```python
df = pd.read_csv('mnist.csv')
```

```python
df.head()
```

```python
print("shape =", df.shape ,", number of dimensions = ", df.ndim)
```

This shows that this is a 2D matrix and it has 42000 training data points with 784 pixels / features with 1 label.

```python
df.columns
```

Above, we can see that the first column is the label column.

```python
labels = df['label']
print(type(labels))
```

```python
# Drop the label column from the dataframe
df = df.drop(['label'], axis=1)
```

```python
#Let's visualize a  sample image
sample_image = df.iloc[1].as_matrix().reshape(28, 28)
plt.imshow(sample_image, cmap='gray')
plt.show()
```

Lets visuaize the data after transforming it to a 2D data

```python
print("shape of the data is", df.shape)
print("shape of the labels is", labels.shape)
print(type(df))
```

```python
#Now let's standardize the data so as to make its
#mean = 0 and standard deviation = 1
sc = StandardScaler()
data = sc.fit_transform(df)
print(data.shape)
print(type(data))
```

```python
#Now lets calculate the covariance matrix
cov_mat = (1/data.shape[0]) * np.dot(data.T, data)
print(cov_mat.shape)
```

```python
#Get the eigen vectors of the largest 2 eigen values because we want to project the data on 2 dimensions.
eigval, eigvec = eigh(cov_mat, eigvals=(782, 783))
```

```python
#These are the list of the top 2 eigen values
eigval
```

```python
eigvec.shape
```

```python
#Now lets project the data points onto the 2Dimensions.
projected_data = np.dot(data, eigvec)
print(projected_data.shape)
```

```python
df = pd.DataFrame(projected_data, columns=['1st principal component', '2nd principal component'])
df['label'] = labels
```

```python
# This is the projected data on 2 dimensions
print(type(df))
df
```

```python
#Lets plot to see what PCA can do
sns.FacetGrid(df, hue='label', size=10) \
    .map(plt.scatter, '1st principal component', '2nd principal component') \
    .add_legend()
plt.title('Scatter plot between 1st and 2nd principal components')
plt.show()
```

# From the above plot, you can say that although PCA helps in reducing the data to lower dimesions it does not separate the classes, visualizing and giving decisions based on this can be quite difficult. So, for tasks where you want to visualize data properly and have points projected onto smaller dimensions while keeping the classes in mind, it will be better to use techniques like LDA or t-SNE
