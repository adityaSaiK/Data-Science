Exploratoy data analysis is the task of understanding the data set through different methods and especially through visual methods.
This should be the first thing with the data. Here we will look into some plotting techniques.

The dataset that we use is called the iris dataset. It contains samples from 3 different flowers. Objective is to classify a new
flower as one of the 3 flower species. So this basically belongs to a classification problem.

The iris dataset has 4 attributrs (independent variables) which we will look into later and 1 class label (dependent variable).
The dataset has 150 observations / data points.

One thing that we have to understand is that it is extremely inportant to have domain knowledge.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
iris = pd.read_csv('Iris.csv', index_col = None)
```

```python
#look at the data type of iris
type(iris)
```

```python
#shape of the iris dataset
iris.shape
```

```python
#removing the Id column
iris = iris.drop(['Id'], axis=1)
```

since all the classes are having same number of points, this dataset is called a balanced dataset.

2D scatter plot

```python
#plot between sepal width and length
plt.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'])
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.title('2D graph between sepal length and sepal width')
plt.show()
```

```python
#let's make it a bit more colorful so that it's a bit easy to understand
sns.FacetGrid(iris, hue="Species", size=6).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()
plt.show()
```

From the above plot you can say the Setosa can easily be separated by drawing a straight line whereas it is difficult to separate Versicolor and Virginica using sepal width and length.

But we are having 4 attributes, so how do we visualize all of them at once ? since we cannot visualize 4 dimesions...

```python
sns.set_style('whitegrid')
sns.pairplot(iris, hue='Species', size=4)
plt.show()
```

By looking at the above graphs, 1) Setosa is really easy to separate. 2) Petal length and Petal Width are important attributes that are able to distinguish the flowers mostly.

Since we have seen bivariate analysis, lets go a step back and work on univariate analysis.

```python
sns.FacetGrid(iris, hue="Species", size=4).map(sns.distplot, "PetalLengthCm").add_legend()
plt.show();
```

```python
sns.FacetGrid(iris, hue="Species", size=4).map(sns.distplot, "PetalWidthCm").add_legend()
plt.show();
```

```python
sns.FacetGrid(iris, hue="Species", size=4).map(sns.distplot, "SepalLengthCm").add_legend()
plt.show();
```

```python
sns.FacetGrid(iris, hue="Species", size=4).map(sns.distplot, "SepalWidthCm").add_legend()
plt.show();
```

By looking at the above PDF's you can say that using Petal length, we can separate Setosa from the other easily whereas with rest of the attributes it's not that easy to draw a boundary that separates them 

The importance of various attributes in separating the classes are Petal_length > Petal_width > Sepal_length > Sepal_width. This was determined by the amount of overlap seen in the respective PDF's.
